```python
#!/usr/bin/env python3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
csv_file_path = '../Dataset/Summary/rf_scores.csv'  # Replace with your actual CSV file path
df = pd.read_csv(csv_file_path)

# Make summary df
rf_summary_df = df.groupby(['folder', 'inner_folder', 'method'])['rf-score'].agg(['mean', 'std']).reset_index()

# Save the summary data to a new CSV file
rf_summary_csv_path = '../Dataset/Summary/rf_scores_summary.csv'  # Replace with your desired output file path
rf_summary_df.to_csv(rf_summary_csv_path, index=False)

# Unique folders
folders = df['folder'].unique()

# Iterate over each folder to create separate plots
for folder in folders:
    # Filter data for the current folder
    folder_data = df[df['folder'] == folder]

    # Create a bar plot for each inner folder/method combination
    plt.figure(figsize=(20, 12))
    sns.barplot(data=folder_data, x='inner_folder', y='rf-score', hue='method', errorbar='sd', palette='tab10')

    # Plot adjustments
    plt.title(f'Average RF Scores in {folder}')
    plt.xlabel('Inner Folder')
    plt.ylabel('Average RF Score')
    plt.xticks(rotation=45)
    plt.legend(title='Method', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()

    # Show the plot
    plt.show()
```
